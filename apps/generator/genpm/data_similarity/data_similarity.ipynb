```python
from genpm.data_similarity.data_similarity_utils import (
    create_spark_session,
    create_demo_dataframes,
    compute_single_metrics,
    compute_multi_metrics,
    create_single_kpi_figure,
    create_multi_kpi_figure,
)
```

```python
spark = create_spark_session()
```

```python

real_sdf, synth_sdf, real_multi_sdf, synth_multi_sdf = create_demo_dataframes(spark)

single_metrics = compute_single_metrics(
    real_sdf=real_sdf,
    synth_sdf=synth_sdf,
    value_col="value",
    ts_col="ts",
    acf_max_lag=24 * 8,
    ls_min_period_h=2.0,
    ls_max_period_h=24 * 14,
    ls_n_freq=2000,
)

multi_metrics = compute_multi_metrics(
    real_sdf=real_multi_sdf,
    synth_sdf=synth_multi_sdf,
    value_cols=["kpi_a", "kpi_b", "kpi_c"],
    ts_col="ts",
    n_projections=200,
)

single_fig = create_single_kpi_figure(
    metrics=single_metrics,
    use_separate_legends=True,
)

multi_fig = create_multi_kpi_figure(
    metrics=multi_metrics,
)
```

```python
single_fig.show()
```
