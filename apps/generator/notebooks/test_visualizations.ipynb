```python
from genpm.raw_vis.data_vis_utils import (
    basic_info,
    kpi_bts_coverage,
    kpi_catalog,
    plot_kpi_timeline,
    schema,
)
```

```python
# use spark session if needed
from genpm.utils.consts import RAW_DATASET_PATH
from genpm.utils.utils import SparkDataManager

sdm = SparkDataManager()


pm_df = sdm.read_parquet(RAW_DATASET_PATH)
```

```python
schema = schema(pm_df)
schema
```

```python
basic_info = basic_info(pm_df)
basic_info
```

```python
kpi_catalog = kpi_catalog(pm_df)
kpi_catalog
```

```python
import plotly.io as pio

heat_json = kpi_bts_coverage(pm_df)
fig = pio.from_json(heat_json)
fig.show()
```

```python
kpis_list = pm_df.select("kpi_id").distinct().rdd.flatMap(lambda x: x).collect()
```

```python
import json

kpi_plots = {}
for kpi in kpis_list[:10]:
    fig = plot_kpi_timeline(pm_df, kpi)
    kpi_plots[kpi] = json.loads(fig.to_json())

analysis = {
    "kpi_list": kpis_list[:10],
    "kpi_plots": kpi_plots,
}
with open("kpi_analysis.json", "w", encoding="utf-8") as f_out:
    json.dump(analysis, f_out, ensure_ascii=False, default=str)
```

```python
for kpi in kpis_list[:10]:
    fig = plot_kpi_timeline(pm_df, kpi)
    fig.show()
```
