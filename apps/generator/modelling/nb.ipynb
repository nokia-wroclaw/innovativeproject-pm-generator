```python
from cvae_utils import (
        prepare_data, build_model, train_model,
        save_artifacts, load_artifacts,
        generate_timespan, generate_n_weeks,
    )

from pathlib import Path
```

```python
SHARED_DIR_PATH = Path(f"/home/{USER}/app/apps/apps/generator/data/shared_dir")
```

```python
TRAINING_DATA_PATH = SHARED_DIR_PATH / "preprocessed_dataset" / "final_scaled_only_minmax"

ARTIFACTS_DIR_PATH = SHARED_DIR_PATH / "artifacts"
RUN_DIR_PATH  = ARTIFACTS_DIR_PATH / "run_3"
WEIGHTS_PATH  = RUN_DIR_PATH / "models_weights"
```

```python
MODEL_PATH  = WEIGHTS_PATH / "cvae_lstm_v4_0.weights.h5"
```

```python
data = prepare_data(TRAINING_DATA_PATH / "pm_df_wide_indexed_winds")
```

```python
X_scaled = data["X_scaled"]
import numpy as np
print(f"X_scaled: mean={X_scaled.mean():.3f}, std={X_scaled.std():.3f}")
print(f"           min={X_scaled.min():.1f}, max={X_scaled.max():.1f}")
per_feat_std = X_scaled.std(axis=(0, 1))
print(f"Per-feature std: min={per_feat_std.min():.2f}, max={per_feat_std.max():.2f}, median={np.median(per_feat_std):.2f}")
print(f"Features with std > 10: {(per_feat_std > 10).sum()} / {len(per_feat_std)}")
```

```python
arch, model = build_model(data["seq_len"], data["feat_dim"], data["output_dim"])
history = train_model(model, data["X_scaled"], data["y_extended"],
                        weights_path=MODEL_PATH)
```

```python
save_artifacts(RUN_DIR_PATH, model, data)
```
